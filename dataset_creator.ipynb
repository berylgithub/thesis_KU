```python
f_euclid_dist = lambda a,b: np.linalg.norm(a-b)

def f_h_step(a, x):
    return 1 if (x<a) else 0

f_y = lambda k : -np.log10(k)
```

```python
'''
======inter-atom-interaction calculator======
- considered atom types for binding (Pedro, Mitchell (2010)) = proteins <-> ligands : {C,N,O,S} <-> {C,N,O,F,P,S,Cl,Br,I} -> N^36
- (optional) complex molecules shall be stored in dictionaries, e.g : {  'name':'2tpi',
                                                            'k':2e-10,
                                                            'protein_atoms': 
                                                                        'C':[[],[],...],
                                                                        'N':[[],[],...],
                                                                        .......
                                                            'ligand_atoms': .......
                                                        }
'''
ID_atoms = ID_ligand = ['C','N','O','F','P','S','Cl','Br','I']
ID_protein = ['C','N','O','S']

def inter_atom_calc(protein_atom, ligand_atom, protein, ligand, cutoff = 12):
    coor=['x','y','z']
    '''get the coordinates of protein atoms r=(x,y,z)'''
    prot_coor = np.array([(protein.loc[protein['atom'] == protein_atom][c_]).values.tolist() for c_ in coor], dtype=float).T
    
    '''coordinates of ligand atoms'''
    lig_coor = np.array([(ligand.loc[ligand['atom'] == ligand_atom][c_]).values.tolist() for c_ in coor], dtype=float).T
    
    #print(len(prot_coor))
    #print(len(lig_coor))
    
    '''calculate euclidian distances between atom and do heavside step function with the cutoff distance (in Armstrong)
        and get the sum
    '''
    sum_inter = 0
    for i in range(len(prot_coor)):
        for j in range(len(lig_coor)):
            sum_inter += f_h_step(cutoff, f_euclid_dist(prot_coor[i], lig_coor[j])) 
    return sum_inter
    
    
```

```python
'''construct the x array (the size should be 4*9)'''
'''
'''
x = np.zeros(len(ID_protein)*len(ID_ligand))
k=0
for p_ in ID_protein:
    for l_ in ID_ligand:
        x[k]=inter_atom_calc(p_, l_, df_protein, df_ligand)
        k+=1
print(x)
```
