```python
import numpy as np
import pandas as pd
import sys, os
```

```python
f_euclid_dist = lambda a,b: np.linalg.norm(a-b)

def f_h_step(a, x):
    return 1 if (x<a) else 0

f_y = lambda k : -np.log10(k)
```

```python
'''set the id of complex'''
id_complex = '2zcq'

'''======y-construction========'''
'''
==Ki/Kd/IC50 data==
this data ONLY changes if the entire dataset is changed, otherwise it stays the same
'''
path = 'C:/Users/user/Documents/Computational Science/Kanazawa/Thesis/Dataset/refined-set/index/INDEX_refined_data.2018'
l = []
with open(path, 'r') as f:
    for line in f:
        if not line.startswith('#'):
            clean_line = line.rstrip()
            l.append((line.rstrip()).split())
df_idx = (pd.DataFrame(l)).rename(columns={0:'id',3:'-log10_k',4:'k'})
print(df_idx)

#########################################
# change the 'string' units to float
set_units = set([d[-2:] for d in df_idx['k']])
print(set_units)
mol_units = {'uM':1.e-6, 'pM':1.e-12, 'fM':1.e-15, 'nM':1.e-9, 'mM':1.e-3}

# get the real values of k
k_values = []
for i, d in enumerate(df_idx['k']):
    k_val = {}
    k_val['id'] = df_idx.iloc[i]['id']
    k_val['k'] = float(d.split('=')[1][:-2])*mol_units[d[-2:]]
    k_val['log10_k'] = f_y(k_val['k'])
    k_values.append(k_val)
print(k_values)
########################################

y = float(df_idx.loc[df_idx['id']==id_complex]['-log10_k'])
print(y)
```

```python
'''ligand data'''
#path='/home/berylramadhian/Documents/thesis_KU/Dataset/pdbbind_v2018_refined/refined-set/'+id_complex+'/'+id_complex+'_ligand.sdf'
path = 'C:/Users/user/Documents/Computational Science/Kanazawa/Thesis/Dataset/refined-set/'+id_complex+'/'+id_complex+'_ligand.sdf'
df_ligand = pd.DataFrame(pd.read_csv(path, delim_whitespace=True, skiprows=[0,1,2,3], usecols=[0,1,2,3], names = ["x","y","z","atom"]))
df_ligand
```

```python
'''protein data'''
#path='/home/berylramadhian/Documents/thesis_KU/Dataset/pdbbind_v2018_refined/refined-set/'+id_complex+'/'+id_complex+'_protein.pdb'
path = 'C:/Users/user/Documents/Computational Science/Kanazawa/Thesis/Dataset/refined-set/'+id_complex+'/'+id_complex+'_protein.pdb'
l = []
with open(path, 'r') as f:
    for line in f:
        if line.startswith('ATOM'):
            clean_line = line.rstrip()
            l.append((line.rstrip()).split())
df_protein = pd.DataFrame(l)
df_protein = df_protein.rename(columns={6:'x', 7:'y', 8:'z', 11:'atom'})
df_protein
```

```python
'''
======x-array construction======
- considered atom types for binding (Pedro, Mitchell (2010)) = proteins <-> ligands : {C,N,O,S} <-> {C,N,O,F,P,S,Cl,Br,I} -> N^36
- (optional) complex molecules shall be stored in dictionaries, e.g : {  'name':'2tpi',
                                                            'k':2e-10,
                                                            'protein_atoms': 
                                                                        'C':[[],[],...],
                                                                        'N':[[],[],...],
                                                                        .......
                                                            'ligand_atoms': .......
                                                        }
'''
ID_atoms = ID_ligand = ['C','N','O','F','P','S','Cl','Br','I']
ID_protein = ['C','N','O','S']


def inter_atom_calc(protein_atom, ligand_atom, protein, ligand, cutoff = 12):
    coor=['x','y','z']
    '''get the coordinates of protein atoms r=(x,y,z)'''
    prot_coor = np.array([(protein.loc[protein['atom'] == protein_atom][c_]).values.tolist() for c_ in coor], dtype=float).T
    
    '''coordinates of ligand atoms'''
    lig_coor = np.array([(ligand.loc[ligand['atom'] == ligand_atom][c_]).values.tolist() for c_ in coor], dtype=float).T
    
    #print(len(prot_coor))
    #print(len(lig_coor))
    
    '''calculate euclidian distances between atom and do heavside step function with the cutoff distance (in Armstrong)
        and get the sum
    '''
    sum_inter = 0
    for i in range(len(prot_coor)):
        for j in range(len(lig_coor)):
            sum_inter += f_h_step(cutoff, f_euclid_dist(prot_coor[i], lig_coor[j])) 
    return sum_inter
    
    
'''construct the x array (the size should be 4*9)'''
'''
'''
x = np.zeros(len(ID_protein)*len(ID_ligand))
k=0
for p_ in ID_protein:
    for l_ in ID_ligand:
        x[k]=inter_atom_calc(p_, l_, df_protein, df_ligand)
        k+=1
print(x)
```
